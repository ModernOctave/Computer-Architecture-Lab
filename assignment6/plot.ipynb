```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from glob import glob

files = glob("submissions/*.observedstat")

!make testzip

names = []
ipcs = []

for file in files:
	name = file.split("/")[-1].split(".")[0]
	with open(file) as f:
		IPC = float(f.readlines()[0].strip())
	names.append(name)
	ipcs.append(IPC)

print(names)
print(ipcs)
```
